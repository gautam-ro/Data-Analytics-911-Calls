### Creating New Columns

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
df
```

```python
df['Reason'].value_counts()
```

### Countplot of 911 calls by reason

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x='Reason', data=df ,palette='bright')
plt.title('911 Calls by Reason')
plt.xlabel('Reason')
plt.ylabel('Count')
plt.legend(labels='Reason')
plt.show()
```

### Conversion of necesarry data

```python
#Converting String to Datetime Data Set
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
#Grabbing particular element i.e; hour,month etc
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
#Converting Days 
day_map = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
df['Day of Week'] = df['Day of Week'].map(day_map)
```

```python
df
```

### Countplot of Day of the week and Month

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20,6))
sns.countplot(x='Day of Week', data=df, hue='Reason', palette='magma')
plt.title('911 Calls by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Count')
plt.legend(loc='upper right')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20,6))
sns.countplot(x='Month', data=df, hue='Reason', palette='magma')
plt.title('911 Calls by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.legend(loc='upper right')
plt.show()
```

### Missing Months

```python
byMonth = df.groupby('Month').count()
df
```

### Plots for this data

```python
plt.figure(figsize=(12, 6))
byMonth['e'].plot(marker='o', linestyle='-', color='b')
plt.title('911 Calls per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
#lmplot() to create a linear fit on the number of calls per month
sns.lmplot(x='Month', y='e', data=byMonth.reset_index(), aspect=1, height=6)
plt.title('Linear Fit of 911 Calls per Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

## Heat Map & Cluster Map for Day of Week

```python
restructured_data = df.pivot_table(values='e', index='Day of Week', columns='Hour', aggfunc='count')
restructured_data
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(restructured_data, cmap='viridis')
plt.title('911 Calls Heatmap')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.clustermap(restructured_data, cmap='viridis')
plt.title('911 Calls Clustered Heatmap')
plt.show()
```
